# Deploy Hugging Face Full Stack Space

Este notebook prepara y revisa el bundle del Docker Space full stack para Turbobujias.

Objetivo:
- generar `dist/hf-fullstack-space`
- validar que el bundle tenga los archivos esperados
- dejar listo el contenido que debes subir a un Docker Space en Hugging Face

```python
from pathlib import
Path
import json
import shutil
import subprocess

repo_root = Path
(r"c:\Users\ASUS\tbai\Turbobujias-pro")
deploy_dir = repo_root / "deploy" / "huggingface-fullstack-space"
dist_dir = repo_root / "dist" / "hf-fullstack-space"
prepare_script = repo_root / "scripts" / "prepare-hf-fullstack-space.mjs"

print({
    "repo_root": str
(repo_root),
    "deploy_dir_exists": deploy_dir.exists
(),
    "prepare_script_exists": prepare_script.exists
(),
    "dist_dir_exists": dist_dir.exists
(),
})
```

```python
result
= subprocess.run
(
    ["node", str(prepare_script)],
    cwd=repo_root,
    capture_output=True,
    text=True,
    check=False,
)
print(result.stdout)
if result.stderr:
print(result.stderr)
print({"returncode": result.returncode
})
```

```python
expected_paths
= [
    dist_dir / "Dockerfile",
    dist_dir / "README.md",
    dist_dir / "backend",
    dist_dir / "turbobujias-web",
    dist_dir / "turbobujias-ai",
    dist_dir / "docker",
]

status = {str
(path.relative_to
(dist_dir)): path.exists
() for path in expected_paths}
status
```

```python
readme_text
=
(dist_dir / "README.md").read_text
(encoding="utf-8")
print(readme_text
[:1200])
```

```python
zip_base
= repo_root / "dist" / "hf-fullstack-space"
zip_path = shutil.make_archive
(str
(zip_base), "zip", root_dir=dist_dir)
print({"zip_path": zip_path})
```

## Despliegue manual en Hugging Face

1. Crea un nuevo **Docker Space** en Hugging Face.
2. Sube el contenido de `dist/hf-fullstack-space/` o el ZIP generado.
3. Configura los secrets necesarios, por ejemplo: `GITHUB_TOKEN`, `GEMINI_API_KEY`, `HF_TOKEN`, `PAYPAL_CLIENT_ID`, `PAYPAL_CLIENT_SECRET`, `MERCADOPAGO_ACCESS_TOKEN`.
4. Espera el build del Space.
5. Valida estas rutas públicas: `/`, `/api/health`, `/api/config/public`, `/api/ai-chat`, `/chatbot/`, `/openapi.json`.
